```python
import matplotlib.pylab as plt
```

```python
import numpy as np
import mdtraj as md
```

```python
t = md.load('/home/teliu/work/aspp2/amber/pp53l-iaspp/1/image-test/fitted.xtc', top='/home/teliu/work/aspp2/amber/pp53l-iaspp/1/image-test/tleapdone.pdb')
```

```python
t.xyz.shape
```

```python
print(t.xyz)
```

```python
top = t.topology
```

```python
selection = top.select('resid 0 to 268 and name CA')
```

```python
lig = top.select('resid 269 to 433 and name CA')
```

```python
print (selection)
```

```python
xcord = t.xyz[:,selection,0]
```

```python
ycord = t.xyz[:,selection,1]
```

```python
allx=np.ravel(xcord)
```

```python
ally=np.ravel(ycord)
```

```python
lig_x = t.xyz[:,lig,0]
lig_y = t.xyz[:,lig,1]
```

```python
xaverage = np.mean(xcord,axis=0)
```

```python
#314 358  zxy
plt.figure(figsize=(3,3))
for i in range(100):
    plt.plot(xcord[i,0:232],ycord[i,0:232],color='cyan',alpha=0.2,lw=1)#cyan
    plt.plot(lig_x[i,0:110],lig_y[i,0:110],color='#7fff7f',alpha=0.3,lw=1)#green
    plt.plot(lig_x[i,126:166],lig_y[i,126:166],color='#7fff7f',alpha=0.3,lw=1)#green
    plt.plot(xcord[i,233:268],ycord[i,233:268],color='#ffff7f',alpha=0.7,lw=1)#yellow
    plt.plot(lig_x[i,111:125],lig_y[i,111:125],color='pink',alpha=0.7,lw=1)#red
plt.plot(np.mean(xcord,axis=0),np.mean(ycord,axis=0),color='gray',lw=1,alpha=0.7)
plt.plot(np.mean(lig_x,axis=0),np.mean(lig_y,axis=0),color='gray',lw=1,alpha=0.7)
plt.xticks(np.arange(-4,10,2),fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('X Axis',fontsize=10)
plt.ylabel('Y Axis',fontsize=10)
plt.title('6RZ3',fontsize=10)

plt.xlim([-4,8])
plt.ylim([-4,6])

plt.savefig('/home/teliu/work/aspp2/figure/fig2/6rz3.png',bbox_inches='tight',transparent=True)
```
